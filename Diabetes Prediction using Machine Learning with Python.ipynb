**`=> importing the dependencies`**

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn import svm
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

**`=> data collection and analysis`**

```python
#loading the dataset
df = pd.read_csv("diabetes.csv")
```

```python
# printing first two rows of data
df.head(2)
```

```python
# number of rows and columns in this dataset
df.shape
```

```python
# getting the statistical meashre of data
df.describe()
```

```python
# checking there are null values or not?
df.isnull().sum()
```

```python
df["Outcome"].value_counts()
```

**`0 => non-diabetic,
1 => diabetic`**

```python
df.groupby("Outcome").mean()
```

```python
# seprate data and labels 
X= df.drop(columns= "Outcome", axis= 1)
Y= df["Outcome"]
```

print(X)

```python
print(Y)
```

**`=> standarized data`**

```python
scaler = StandardScaler()
standarized_data = scaler.fit_transform(X)
```

```python
print(standarized_data)
```

```python
X = standarized_data
Y = df["Outcome"]
```

```python
print(X)
print(Y)
```

**`=> spliting the data into train and test`**

```python
X_train, X_test, Y_train, Y_test= train_test_split(X, Y, random_state= 38, test_size= 0.2, stratify= Y)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

**`=> training our model`**

```python
classifier= svm.SVC(kernel= "linear")
classifier.fit(X_train, Y_train)
```

**`=> model evaluation`**

```python
# accuracy score on training data
X_train_prediction= classifier.predict(X_train)
accuracy_training_data= accuracy_score(X_train_prediction, Y_train)
```

```python
print(accuracy_training_data)
```

```python
# accuracy score on testing data
X_test_pred= classifier.predict(X_test)
accuracy_testing_data= accuracy_score(X_test_pred, Y_test)
```

```python
print(accuracy_testing_data)
```

 **`=> making prediction system`**

```python
input_data =(8,99,84,0,0,35.4,0.388,50)

# convert input_data into numpy array
input_data_as_numpy_array = np.array(input_data)

# reshape the input_data_as_numpy_array
input_data_reshaped= input_data_as_numpy_array.reshape(1, -1)

# standrized the input data
std_data= scaler.transform(input_data_reshaped)
print(std_data)

prediction = classifier.predict(std_data)
print(prediction)

if prediction[0] == 0:
    print("Patient is Non-Diabetics")
else:
    print("Patient is Diabetic")
```
